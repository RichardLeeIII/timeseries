## Importing the relevant packages


AR model only works good on stationary data, it works poorly on non-stationary data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.arima.model import ARIMA 
from scipy.stats.distributions import chi2
import statsmodels.tsa.stattools as sts 
import seaborn as sns
sns.set()
```

## Importing the Data and Pre-processing 

```python
raw_csv_data = pd.read_csv("Index2018.csv") 
df_comp=raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.set_index("date", inplace=True)
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

```python
df_comp['market_value']=df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## The ACF

```python
sgt.plot_acf(df.market_value, zero = False, lags = 40)
plt.title("ACF for Prices", size = 20)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

## The PACF

```python
sgt.plot_pacf(df.market_value, lags = 40, alpha = 0.05, zero = False, method = ('ols'))
plt.title("PACF for Prices", size = 20)
plt.show()
```

## The AR(1) Model

```python
model_ar = ARIMA(df.market_value, order=(1,0,0))  # ARIMA with order (p, d, q), (1, 0, 0) is equivalent to AR(1) p=lag, d=trend and seasonality, q=Moving average, including residuals
```

```python
results_ar = model_ar.fit()
```

```python
results_ar.summary()
```

```python
#P>|z| for ar.L1 is essentially 0: This means the null hypothesis (that the coefficient is zero) 
#can be rejected with a very high level of confidence. 
#The coefficient is significantly different from zero.
```

## LLR Test

```python
def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR, DF).round(3)
    return p
```

## Comparing Higher-Lag AR Models

```python
LLR_test(model_ar_2, model_ar_3)
```

```python
LLR_test(model_ar_3, model_ar_4)
```

```python
model_ar_4 = ARIMA(df.market_value, order=[4,0,0])
results_ar_4 = model_ar_4.fit()
print(results_ar_4.summary()) 
print ("LLR test: " + str(LLR_test(model_ar_3, model_ar_4)))
```
